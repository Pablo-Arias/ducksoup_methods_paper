This notebook analyses the data that was formated in MARS

```python
import glob
import polars as pl
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pyarrow
```

## Mean pitch

```python
#This data is already collapsed, is not the time series
al_data_df = pd.read_csv("../data/audio_data/collapsed_pitch_from_server.csv")
```

## Plot

```python
g = sns.catplot( x="participant_condition", y="f0_mean", units="user_id"
                    , height=4.6
                    , aspect=0.8
                    , data=al_data_df
                    , kind="point"
                    , capsize=.16
                    , hue="other_condition"
                    , legend=False
                    , scale=1
                    , dodge= 0.15
                    , errwidth=2
                    , order=["S", "U"]
                    , hue_order=["S", "U"]
                    , palette= sns.color_palette(["red", "blue"])
            )

# Add individual measure points
sns.stripplot(x="participant_condition", y="f0_mean"
                , data=al_data_df
                , hue="other_condition", dodge=0.15, size=5, alpha=0.3
                , legend=False
                , order=["S", "U"]
                , hue_order=["S", "U"]
                , palette= sns.color_palette(["red", "blue"])
                )    
    
#g.set(ylim=(0, 1.7))

plt.savefig("plots/mean_pitch_legend.pdf", bbox_inches='tight')
g.set(xticklabels=[], yticklabels=[], title='',xlabel=None, ylabel=None)
plt.savefig("plots/mean_pitch.pdf", bbox_inches='tight')
```

## STD

```python
# STANDARD DEVIATION

g = sns.catplot( x="participant_condition", y="f0_std", units="user_id"
                    , height=4.6
                    , aspect=0.8
                    , data=al_data_df
                    , kind="point"
                    , capsize=.16
                    , hue="other_condition"
                    , legend=False
                    , scale=1.1
                    , dodge= 0.15
                    , errwidth=2
                    , order=["S", "U"]
                    , hue_order=["S", "U"]
                    , palette= sns.color_palette(["red", "blue"])
            )

# Add individual measure points
sns.stripplot(x="participant_condition"
                , y="f0_std"
                , data=al_data_df
                , hue="other_condition", dodge=0.15, size=5, alpha=0.3
                , legend=False
                , order=["S", "U"]
                , hue_order=["S", "U"]
                , palette= sns.color_palette(["red", "blue"])
                )    
    
#g.set(ylim=(0, 1.7))

plt.savefig("plots/std_pitch_legend.pdf", bbox_inches='tight')
g.set(xticklabels=[], yticklabels=[], title='', xlabel=None, ylabel=None)
plt.savefig("plots/std_pitch.pdf", bbox_inches='tight')
```
